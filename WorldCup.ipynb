```python
# 统一"Germany FR"和"Germany"
summary = summary.replace(['Germany FR'],'Germany')
```

```python
# 历年现场观众人数变化趋势
plt.figure(figsize = (16,10))
plt.plot(summary['Year'],summary['Attendance'],marker='o',linestyle = '-',color = 'b')
plt.title("Attendance's Tendency")
plt.xlabel('Years')
plt.ylabel('Attandance')
plt.xticks(summary['Year'])#设置x轴刻度
plt.grid(True)#开启网格线
plt.tight_layout()#自动调整图表元素的布局，以避免重叠

plt.show()
```

```python
# 参赛队伍数变化趋势
plt.figure(figsize = (16,10))
plt.plot(summary['Year'],summary['QualifiedTeams'],marker='o',linestyle = '-',color = 'b')
plt.title("QualifiedTeams' Tendency")
plt.xlabel('Years')
plt.ylabel('QualifiedTeams')
plt.xticks(summary['Year'])
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# 历年进球数变化趋势
plt.figure(figsize = (16,10))
plt.plot(summary['Year'],summary['GoalsScored'],marker='o',linestyle = '-',color = 'b')
plt.title("GoalsScored's Tendency")
plt.xlabel('Years')
plt.ylabel('GoalsScored')
plt.xticks(summary['Year'])
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# 夺冠次数分析
winner_counts = summary['Winner'].value_counts()
plt.figure(figsize=(16,10))
winner_counts.plot(kind='bar',color='skyblue')
plt.title('Number of Championships')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # 0度表示横着显示
plt.grid(axis='y', linestyle='--', alpha=0.7)# 显示网格
plt.tight_layout()

plt.show()
```

```python
# 半决赛（4强）队伍次数统计
countries = summary[['Winner','Second','Third','Fourth']].apply(lambda x:pd.Series(x).value_counts()).reset_index().fillna(0)
countries['SemiFinal'] = countries['Winner']+countries['Second']+countries['Third']+countries['Fourth']
countries['Final'] = countries['Winner']+countries['Second']
countries
```

```python
colors = ['blue' if (i>=8) else 'yellow' if(5<=i<8) else 'purple' if (3<=i<5) else 'orange' if (i==2) else 'red' for i in countries['SemiFinal']]

plt.figure(figsize=(16,10))
sns.barplot(data=countries,x='index',y='SemiFinal',hue='index',palette=colors,legend=False)
plt.title('SemiFinal Statistic')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45,ha='center',fontsize=12) 
plt.tight_layout()
plt.show()
```

```python
# 决赛队伍次数统计
FinalList = countries.drop(countries[countries['Final']==0].index)
clrs= ['blue' if (i>=6) else 'y' if (i==5) else 'yellow' if (3<=i<5) else 'purple' if (i==2) else 'red' for i in FinalList['Final']]
plt.figure(figsize=(16,10))
sns.barplot(data=FinalList,x='index',y='Final',hue='index',palette=clrs,legend=False)
plt.title('Final Statistic')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45,ha='center',fontsize=14) 
plt.tight_layout()
plt.show()
```

```python
# 进入决赛后夺冠以来分析
FinalList['ratio'] = FinalList['Winner']/FinalList['Final']
plt.figure(figsize=(16,10))
sns.barplot(data=FinalList,x='index',y='ratio',hue='index',legend=False)
plt.title('Percentage of winning reaching the final')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45,ha='center',fontsize=14) 
plt.tight_layout()
plt.show()
```

```python
# 夺冠队伍所在大洲分布
index1 = summary['WinnerContinent'].value_counts().index.tolist()
value1 = summary['WinnerContinent'].value_counts().values.tolist()

palette = ['skyblue','green']

plt.pie(value1,labels=index1,autopct='%.0f%%',colors=palette,wedgeprops={'edgecolor':'0','linewidth':2.5})
plt.title('Champion Continent Ratios')
```

```python
# 东道主进入半决赛/决赛/夺冠概率统计
# .any(axis=1)用于行级别的布尔判断
# .eq 是一个用于比较的函数，通常在 Pandas 中用于返回一个布尔数组，表示每个元素是否等于给定的值
# 东道主是否进入四强 决赛 夺冠
summary['HostTop4'] = summary[['Winner','Second','Third','Fourth']].eq(summary['HostCountry'],axis=0).any(axis=1)
summary['HostTop2'] = summary[['Winner','Second']].eq(summary['HostCountry'],axis=0).any(axis=1)
summary['Hostwinner'] = summary[['Winner']].eq(summary['HostCountry'],axis=0).any(axis=1)
```

```python
total_host = len(summary)
prob_top4 = summary['HostTop4'].sum()/total_host
prob_top2 = summary['HostTop2'].sum()/total_host
prob_winner = summary['Hostwinner'].sum()/total_host

probabilities = [prob_top4,prob_top2,prob_winner]
labels = ['Top4', 'Top2', 'Winner']

plt.figure(figsize=(16,10))
barplot = sns.barplot(x=labels,y=probabilities,hue=labels,palette='viridis')

# 在柱形图顶端显示数据
for index, value in enumerate(probabilities):
    barplot.text(index, value, f'{value:.2f}', ha='center', va='bottom', fontsize=12)

plt.title('The probability of the host at different stages')
plt.ylabel('Porobability')
plt.ylim(0,1)# 设置 y 轴范围为 0 到 1
plt.grid(False)

plt.show()
```

```python
# 分析世界杯比赛信息表
matches = pd.read_csv('WorldCupMatches.csv')
```

```python
# 数据预处理
matches = matches.replace(['Germany FR'],'Germany')

# 类型转化
matches['Home Team Goals']= matches['Home Team Goals'].astype(int)
matches['Away Team Goals']= matches['Away Team Goals'].astype(int)

# 格式化比赛结果，如 3-2
matches['result'] = matches['Home Team Goals'].astype(str)+"-"+matches['Away Team Goals'].astype(str)
matches
```

```python
# 现场观赛人数分析
# 筛选出热门比赛
top5_attendance = matches.sort_values(by='Attendance',ascending=False)[:5]
top5_attendance

top5_attendance['VS'] = top5_attendance['Home Team Name'] + " "+ "VS "+ " " + top5_attendance['Away Team Name']

plt.figure(figsize=(16,10))
ax = sns.barplot(y=top5_attendance['VS'],x=top5_attendance['Attendance'],hue=top5_attendance['Attendance'],palette='viridis')

plt.title('Top 5 Matches')
plt.ylabel('Teams')
plt.xlabel('Attandance')

for i,s in enumerate("Stadium:"+top5_attendance['Stadium']+"\nDate:"+top5_attendance['Datetime']+"\nAttandance"+
                     top5_attendance['Attendance'].astype(str)):
    ax.text(2000,i,s,fontsize=14,color='white')
    
plt.show()
```

```python
# 比赛进球数分析
matches['VS'] = matches['Home Team Name']+" " + "VS" + " " + matches['Away Team Name']
matches['Total Goals'] = matches['Home Team Goals']+matches['Away Team Goals']

top10_goals = matches.sort_values(by='Total Goals',ascending=False)[:10]
top10_goals['total_goals_str'] = top10_goals['Total Goals'].astype(str)+" goals scored"
top10_goals
plt.figure(figsize=(16,10))
ax = sns.barplot(y=top10_goals['VS'],x=top10_goals['Total Goals'],hue=top10_goals['Total Goals'],palette='viridis')
plt.title('Top 10 Goals Match')
plt.ylabel('Matches')
plt.xlabel('Goals')

for i,s in enumerate("Stadium:"+top10_goals['Stadium']+"\nDate:"+top10_goals['Datetime']+"\n"
                     +top10_goals['total_goals_str']+"Match result:"
                     +top10_goals['result']):
    ax.text(0.2,i,s,fontsize=12,color='white',va = 'center')
plt.show()
```

```python
# 分差最大的比赛
matches['difference_goals'] = pd.Series.abs(matches['Home Team Goals']-matches['Away Team Goals'])

top10_difference = matches.sort_values(by='difference_goals',ascending=False)[:10]
top10_difference['diff_goals'] = top10_difference['difference_goals'].astype(str)+" goals difference"
top10_difference

plt.figure(figsize=(16,10))
ax = sns.barplot(y=top10_difference['VS'],x=top10_difference['difference_goals'],hue=top10_difference['difference_goals'])

plt.title('Top 10 Biggest Difference')
plt.xlabel('Difference')
plt.ylabel('Matches')

for i,s in enumerate("Stadium:"+top10_difference['Stadium']+",Date:"+top10_difference['Datetime']
                     +"stage:"+top10_difference['Stage']+"\n"+top10_difference['diff_goals']+"match result:"+top10_difference['VS']):
    ax.text(0.2,i,s,fontsize = 12,color = 'white',va = 'center')

plt.show()
```
